### Using for loops to call indexes in two_dim_array as i

```python
## calling two_dim_array[i] will yield the ith row

for i in range(len(two_dim_array)):
    print(i)
    print(two_dim_array[i])
```

### passing the row to a second for loop will cycle through the second row
### index as $j$. Calling row [j] will yield the jth value from the row

```python
for i in range(len(two_dim_array)):
    print("i", i)
    print(two_dim_array[i])
    for j in range(len(two_dim_array[i])):
        print("j", j)
        print(two_dim_array[i][j])
```

```python
print("i j:  Val")
for i in range(len(two_dim_array)):
    row = two_dim_array[i]
    for j in range(len(row)):
        print(i, j,"  ", row[j])
```

```python
for row in two_dim_array:
    print(row)
    for val in row:
        print(val)
```

# Create artificial Data

```python
macro_dict = {"GDP": {},
              "Money": {},
              "Real GDP": {},
              "Price Level": {}}
for key in macro_dict.keys():
    for i in range(1990, 2021):
        macro_dict[key][i] = np.random.random() * 10000
macro_dict
```

```python
macro_DF = pd.DataFrame(macro_dict)
macro_DF
```

```python
macro_DF[["GDP"]]
```

```python
# Create a new variable based on mathing existing variables (Velocity)
macro_DF["Velocity"] = macro_DF["Price Level"].mul(macro_DF["Real GDP"].div(macro_DF["Money"]))
macro_DF
```

## matplotlib

```python
import matplotlib.pyplot as plt
macro_DF.plot()
```

```python
# plot using method for matplotlib from the pandas DataFrame
fig, ax = plt.subplots(figsize = (20,12))
macro_DF.plot(linewidth = 5, alpha = .7, legend = False, ax = ax) #Color = "C0" . . .
ax.legend(fontsize = 20)
```

```python
for key in macro_DF:
    fit, ax = plt.subplots(figsize = (10,6))
    macro_DF[key].plot.line(ax = ax)
    ax.set_title(key, fontsize = 20)
```

```python
# Goal is to create multi-plot visualization with seperate plot areas for 
# each variable

# count num variables
num_vars = len(macro_DF.keys())

# create a plot with a subplot for each variable (total 5)
fig, axs = plt.subplots(1, num_vars, figsize = (20,6))

# to plot each variable on each plot, cycle through index of the keys
for i in range(num_vars):
    
    ax = axs[i]
    # call key by index number
    key = list(macro_DF.keys())[i]
    # set color in light of number [i] so that it is Ci
    color = "C" + str(i)
    
    alpha = .2 + .15 * i
    # plot key according to index number in ax
    macro_DF[key].plot.line(color = color, alpha = alpha, ax = ax)
    ax.set_title(key, color = color, fontsize = 20)
```
